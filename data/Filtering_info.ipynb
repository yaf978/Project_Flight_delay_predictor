```python
import pandas as pd
import numpy as np
data = "flights.csv"
flight_data = pd.read_csv(data)
flight_data.head()
```

```python
fixed_flight=flight_data[['MONTH','DAY','AIRLINE','ORIGIN_AIRPORT','DESTINATION_AIRPORT','SCHEDULED_ARRIVAL','ARRIVAL_DELAY']]
fixed_flight
```

```python

for column in fixed_flight.columns:
    print(f"Column {column} has {fixed_flight[column].isnull().sum()} null values")
```

```python
fixed_flight.dropna()
```

```python
def categorise(row):
    if row['ARRIVAL_DELAY'] > 0:
        return '1'
    return '0'
fixed_flight['DELAYED'] = fixed_flight.apply(lambda row: categorise(row), axis=1)
```

```python
fixed_flight
```

```python
fixing_flight = fixed_flight['MONTH'] <= 6
```

```python
final_fix=fixed_flight[fixing_flight]
final_fix
```

```python
final_fix.to_csv('Fixed_flight.csv')
```
